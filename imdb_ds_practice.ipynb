#using apply() function

```python
def rating_category(rating):

  if rating >= 8:
    print('Superb');
  elif (rating >= 7) and (rating < 8):
    print('Good')
  elif (rating >= 5) and (rating < 7):
    print('Moderate')
  else:
    print('Medicore')
   
```

```python
df['Categorical_ratng'] = df['Rating'].apply(rating_category)
```
